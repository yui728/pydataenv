# 4.3 Matplotlib

## 4.3.1 Matplotlibの概要

### Matplotlibとは
* Pythonで主に2次元のグラフを描画するためのライブラリ

Matplotlibのインポート

```python
import matplotlib.pyplot as plt
```

ggplotのインポート

```python
import matplotlib.style
# ggplotスタイルを指定
matplotlib.style.use('ggplot')
```

### MATLABスタイル
* MATLAB（数値解析ソフトウェア）と似た形式でグラフ描写を行う

#### matplotlib.pyplot.plot関数
* X軸、Y軸に設定するデータを設定する

#### matplotlib.pyplot.title関数
* グラフのタイトルを設定する

#### matplotlib.puplot.show関数
* グラフを表示する

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y) # 折れ線グラフを描画
plt.title('MATLAB-style') # グラフにタイトルを設定

plt.show() # グラフを表示
```

### オブジェクト指向スタイル
* 描画オブジェクトに対してサブプロットを追加し、サブプロットに対してグラフを描画する
* 一つの描画オブジェクトに対して、複数のサブプロットを指定することで、複数のグラフを描画可能

#### matplotlib.pyplot.subplots関数
* 描画オブジェクトとサブプロットを生成する
* 第1引数に数値を渡すと、第1引数を行数とした1列のサブプロットが生成される
* 第1引数、第2引数に数値を渡すと、第1引数を行数、第2引数を列数としたサブプロットが生成される
* ncolsキーワード：列数を指定する
* nrowsキーワード：行数を指定する

```python
# データを用意
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト（fig）とサブプロット（ax）を生成
fig, ax = plt.subplots()

ax.plot(x, y) # 折れ線グラフを描画
ax.set_title('OOP-style') # タイトルを設定

plt.show() # グラフを表示
```

## 4.3.3 グラフの種類と出力方法

### 折れ線グラフ

### matlibplot.pyplot.plotメソッド
* 第1引数をX軸、第2引数をY軸にプロットして折れ線グラフを描画する

2本の折れ線グラフを描画する

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 2]
ax.plot(x, y1) # 折れ線グラフを描画
ax.plot(x, y2)

plt.show()
```

*値の間隔を細かくすることで、擬似的に曲線のグラフを作成できる*

sin, cosのグラフを描画する

```python
import numpy as np

x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig,ax = plt.subplots()
ax.plot(x, y1, label = 'sin')
ax.plot(x, y2, label = 'cox')
ax.legend()

plt.show()
```

### 棒グラフ

#### axes.Axes.barメソッド
* 第1引数をX軸、第2引数をY軸にとった縦方向の棒グラフを描画する
* 第1引数、第2引数に配列を渡した場合は、対応する要素をX軸、Y軸にとって複数の棒グラフを描画する
* tick_labelキーワード：引数で渡された文字列をメモリへ任意のラベルとして設定する
  * 引数に文字列の配列を渡した場合は、対応する位置の棒グラフのラベルに設定する

シンプルな棒グラフを描画

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]

ax.bar(x, y)

plt.show()
```

ラベル付きの棒グラフを描画する

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]

labels = ['spam', 'ham', 'egg']
ax.bar(x, y, tick_label = labels)

plt.show()
```

#### axes.Axes.barhメソッド
* 第1引数をX軸、第2引数をY軸にとった横向きの棒グラフを描画する

横向きの棒グラフを描画する

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]

labels = ['spam', 'ham', 'egg']
ax.barh(x, y, tick_label = labels)

plt.show()
```

棒グラフを2種類並べて描画する

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3 ,6]
labels = ['spam', 'ham', 'egg']

width = 0.4 # 棒グラフの幅を0.4にする
ax.bar(x, y1, width = width, tick_label = labels, label = 'y1') # 幅を指定して描画

# 幅分ずらして棒グラフを描画
x2 = [num + width for num in x]
ax.bar(x2, y2, width = width, label = 'y2')

ax.legend()

plt.show()
```

積み上げ棒グラフを描画する

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

# y1とy2を足した値を格納
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

# y1とy2を足した高さの棒グラフを描画
ax.bar(x, y_total, tick_label = labels, label = 'y1')
ax.bar(x, y2, label = 'y2')
ax.legend()

plt.show()      
```

様々なマーカーの散布図を描画

```python
fig, ax = plt.subplots()

# 乱打mjに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x[0:10], y[0:10], marker = 'v', label = 'triangle down') # 下向き三角
ax.scatter(x[10:20], y[10:20], marker = '^', label = 'triangle up') # 上向き三角
ax.scatter(x[20:30], y[20:30], marker = 's', label = 'square') # 正方形
ax.scatter(x[30:40], y[30:40], marker = '*',label = 'star') # 星型
ax.scatter(x[40:50], y[40:50], marker = 'x', label = 'x') # X

ax.legend()

plt.show()
```

### ヒストグラム

#### axes.Axes.histメソッド
* 第1引数に渡した配列を値としたヒストグラムを描画する
* bins引数
  * 数値でビンの数を設定する
* orientation引数
  * ヒストグラムの描画の向きを設定する
  * horizonal：横向き
* stacked引数
  * ヒストグラムのビンを積み重ねて描画するか
  * True：積み重ねて描画する
* 戻り値
  1. ヒストグラムの各ビンの度数（要素数）
  1. ビンの境界の値
  1. ビンを描画するための情報

ヒストグラムを描画

```python
# データを生成
np.random.seed(123)
mu = 100 # 平均値
sigma = 15 # 標準偏差
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
n, bins, patches = ax.hist(x)

plt.show()
```

histメソッドの戻り値から、度数分布表を出力

```python
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i], bins[i + 1], num))
```

ビンの数を指定したヒストグラムを描画

```python
fig, ax = plt.subplots()

ax.hist(x, bins = 25) # ビンの数を指定して描画

plt.show()
```

横向きのヒストグラムを描画

```python
fig, ax = plt.subplots()

# 横向きのヒストグラムを描画
ax.hist(x, orientation = 'horizontal')

plt.show()
```

ヒストグラムを並べて描画

```python
# データを生成
np.random.seed(123)
mu = 100 # 平均値
x0 = np.random.normal(mu, 20, 1000)

# 異なる標準偏差でデータを生成
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
# 3つのデータのヒストグラムを描画
ax.hist((x0, x1, x2), label = labels)
ax.legend()

plt.show()
```

積み上げたヒストグラムを描画

```python
fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']
# 積み上げたヒストグラムを描画
ax.hist((x0, x1, x2), label = labels, stacked = True)
ax.legend()

plt.show()
```

### 円グラフ

#### axes.Axes.pieメソッド
* 第1引数で渡した数値の配列をもとにした円グラフを描画する
* startangle引数
  * 引数で渡した角度分ずらした位置から円グラフを開始する
* counterclock引数
  * 反時計回りに配置するかどうか
  * False：時計回りに配置する
* shadow引数
  * グラフに影を描写するかどうか
  * True：影を描写する
* autopct引数
  * 値の%表記のフォーマットを指定する
* explode引数
  * 描画する円グラフの各要素がどれだけ円から切り出されるかを数値で指定する
  * 第1引数で渡す円グラフの値と同じ長さの配列で設定する

円グラフを描画する

```python
labels = ['spam', 'ham', 'egg']
x = [10, 3, 1]

fig, ax = plt.subplots()

ax.pie(x, labels = labels) # 円グラフを描画

plt.show()
```

#### axes.Axes.axisメソッド
* 引数でグラフのアスペクト比を設定する
  * equal：アスペクト比を保持する

アスペクト比を保持した円グラフを描画

```python
fig, ax = plt.subplots()

ax.pie(x, labels = labels)
ax.axis('equal') # アスペクト比を保持して描画する

plt.show()
```

上から時計回りに円グラフを描画

```python
fig, ax = plt.subplots()

ax.pie(x, labels = labels, startangle = 90, counterclock = False) # 上から時計回り
ax.axis('equal')

plt.show()
```

影とパーセント表記を追加した円グラフを描画する

```python
fig, ax = plt.subplots()

ax.pie(x, labels = labels, startangle = 90, counterclock = False, shadow = True, autopct = '%1.2f%%') # 影と％表記を追加
ax.axis('equal')

plt.show()
```

hamを切り出した円グラフを描画

```python
explode = [0, 0.2, 0]

fig, ax = plt.subplots()

ax.pie(x, labels = labels,
      startangle = 90, counterclock = False,
      shadow = True, autopct = '%1.2f%%',
      explode = explode) # explodeを指定する
ax.axis('equal')

plt.show()
```

ヒストグラムと近似曲線を組み合わせて描画

```python
np.random.seed(123)
x = np.random.randn(1000) # 正規乱数を生成

fig, ax = plt.subplots()

# ヒストグラムを描画
counts, edges, patches = ax.hist(x, bins = 25)

# 近似曲線に用いる点を求める（ヒストグラムのビンの中点）
x_fit = (edges[:-1] + edges[1:]) / 2
# 近似曲線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit ** 2 / 2) / np.sqrt(2 * np.pi)
ax.plot(x_fit, y)

plt.show()

print(counts)

print(edges)

patches
```

### フォント

#### フォントに関する引数
* size引数
  * 値に渡した数値を、フォントのポイント数として設定する
* weight引数
  * 値に渡した文字列を、フォントの太さとして設定する
    * bold：太字
    * light：細字
    * heavy：boldよりも太字
* family引数
  * 値に渡した文字列を、フォントの種類として設定す
  * フォントの種類は以下の種類が設定可能
    * serif
    * sans-serif
    * cursive
    * fantasy
    * monospace
* fontdict引数
  * 値に渡した辞書で定義したフォントスタイルを一括で設定する

フォントのスタイルを指定

```python
fig, ax = plt.subplots()

ax.set_xlabel('xlabel', family = 'fantasy', size = 20, weight = 'bold')
ax.set_ylabel('ylabel', family = 'cursive', size = 40, weight = 'light')
ax.set_title('graph title', family = 'monospace', size = 35, weight = 'heavy')

plt.show()
```

フォントのスタイルを辞書形式で指定

```python
# フォントのスタイルを辞書で定義
fontdict = {
    'family': 'fantasy',
    'size': 20,
    'weight': 'normal',
}

fig, ax = plt.subplots()

# 辞書形式でフォントのスタイルを指定
ax.set_xlabel('xlabel', fontdict = fontdict)
ax.set_ylabel('ylabel', fontdict = fontdict)
# 個別指定したsizeで上書き可能
ax.set_title('graph title', fontdict = fontdict, size = 40)

plt.show()
```
